# 2. K-Means Clustering

## a. Standardize

**Since many unsupervised learning techniques are distance-based calculations, sclaing is a required step!**

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit_transform(df)
```

```python
df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
df_scaled
```

```python
df.describe()
```

```python
df_scaled.describe()
```

## b. Clustering

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

inertia_values = []
silhouette_scores = []

for k in range(2, 16):
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
    kmeans.fit(df_scaled)
    inertia_values.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(df_scaled, kmeans.labels_, metric='euclidean', sample_size=None))
```

```python
import matplotlib.pyplot as plt

inertia_series = pd.Series(inertia_values, index=range(2, 16))
inertia_series.plot(marker='o')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.title("Number of Clusters vs. Inertia");
```

```python
# I believe k=4 is a decent number of clusters according to this plot.
```

```python
silhouette_series = pd.Series(silhouette_scores, index=range(2,16))
silhouette_series.plot(marker='o')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Number of Clusters vs. Silhouette Score");
```

```python
# According to silhouette score, k=4 is also the right number.
```

```python
kmeans4 = KMeans(n_clusters=4, n_init=10, random_state=42)

kmeans4.fit(df_scaled)
```

```python
from collections import Counter
Counter(kmeans4.labels_)
```

```python
kmeans4.cluster_centers_
```

```python
cluster_centers_4 = pd.DataFrame(kmeans4.cluster_centers_, columns=df_scaled.columns)

plt.figure(figsize=(10, 2))
sns.heatmap(cluster_centers_4, annot=True, cmap="RdBu", fmt=".1f", linewidths=.5);
```

**Interpreting the clusters:**
* cluster 0: Juniors and Research & Development employees
* cluster 1: Sales employees
* cluster 2: Senior employees
* cluster 3: HR employees

# 3. PCA

Visualizing these clusters using PCA.

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca.fit(df_scaled)
```

```python
pca.explained_variance_ratio_
```

```python
pca.components_
```

```python
df.columns
```

**Interpreting the components:**
*  Component 1: higher age + job level + monthly income = more senior
*  Component 2: higher = Research & Dev / lower = Sales

```python
df_scaled_trasnformed = pd.DataFrame(pca.transform(df_scaled), columns=['PC1', 'PC2'])
df_scaled_trasnformed.head()
```

## Overlay the Kmeans cluster colors

```python
sns.scatterplot(x='PC1', y='PC2', data=df_scaled_trasnformed)
plt.xlabel('More Senior -->')
plt.ylabel('<-- Sales     Research & Dev -->');
```

```python
sns.scatterplot(x='PC1',
                y='PC2',
                data=df_scaled_trasnformed, hue=kmeans4.labels_, palette='viridis');
```

This does not seem like a good way to seperate my data since we can obviously see that there are 3 separate clusters, but the plot is showing 4.
Look at cluster 3, we know that it belongs to HR employees. As a result, this clustering seems like a department segmentation.
